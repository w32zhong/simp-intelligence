Shared memory is organized into 32 banks (to match 32 threads in a warp). Each bank is a "checkout lane" for a SRAM that can load or store 4 bytes (they are designed to cope with half-precision floating point data, the most used type) *at a data cycle*.

The formula for bank-memory mapping is: $$\operatorname{bank}(\text{addr}) = \frac{\text{addr}}{4} \mod 32 $$

there would be a *bank conflict* (and they have to be queued) if two requests access the same bank:
![](./banks-conflict.png)

Using a trick called *swizzling*, we can avoid some bank conflicts by changing the layout of the memory access.
