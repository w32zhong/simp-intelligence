## Overall

| Generation       | CUDA Compute Capability                           | Introduced features                                                           |
|------------------|---------------------------------------------------|-------------------------------------------------------------------------------|
| Volta (2018)     | 7.0 (V100)                                        | first-generation tensor cores (FP16)                                          |
| Turing (2019)    | 7.5 (RTX 20xx, Tesla T4)                          | added INT4 and INT8 in tensor cores, warp-level tensor core MMA               |
| Ampere (2020)    | 8.0 (A100) - 8.6 (RTX 30xx, A6000)                | async copy, and adding BF16, TF32, and FP64 in tensor cores -- first NV GPU TF32 adoption, PCIe4 |
| Ada (2023)       | 8.9 (L40S, RTX 6000 Ada, RTX 40xx)                | switched from Samsung 8nm to TSMC 4N fabrication                              |
| Hopper (2022)    | 9.0 (H100, H200)                                  | introduced FP8 (E5M2, E4M3), TMA (global <-> shared), cluster CTAs, PCIe5, HBM3  |
| Blackwell (2024) | 10.0 - 10.3 (B200, B300), 12.0 (RTX 50xx and PRO) | introduced NVFP4, Tensor Memory, and WGMMA (warp semantics) -> tcgen05.mma (async thread semantics) |
| Rubin (2025)     |                                                   | HBM4, more NVFP4 density, and BlueField DPUs for KV-cache storing and high-bandwidth retrieval |
