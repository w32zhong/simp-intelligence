## Note on NVFP4
Compared to FP4 (E2M1), NVFP4 has 4 bits (1 sign, 2 exponent, 1 mantissa) plus 1 shared FP8 scale per 16 value block.

Its predecessor is MXFP4 (plus 1 shared power-of-two scale per 32 value block) which is natively supported by GPT-OSS.
