## Flash Attention Series

* Flash Attention v1 uses [online softmax](./math/online_flash_attn.ipynb) and adds tiling, together solving the Memory IO bottleneck;
* Flash Attention v2 improves GPU thread occupancy with *sequence parallelism* and *loop reordering*;
* Flash Attention v3 targets NVIDIA Hopper (H100) using Warp-specialized producer/consumer to overlap the Softmax calculation with the GEMM of the next block, further improving GPU utilization. Also, it natively supports FP8 quantization.
* [Flash Attention v4](https://github.com/Dao-AILab/flash-attention/blob/5c1627a7a1cda9c32cb9b937a053564e663f81bc/flash_attn/cute/flash_fwd_sm100.py) is implemented in CUTLASS CuTe Python DSL, targeting for Blackwell GPUs. [1]

## FlexAttention [2]

Introduce `score_mod` as an user-defined function, this adds good code flexibility to support

* Relative Position Encodings
* ALiBi Bias
* Soft-capping
* Causal Mask

in addition to full attention.

It also introduces `mask_mod` to achieve sparsity from masking, including

* Causual Attention
* Sliding Window
* PrefixLM (e.g., VLMs)
* Document Masking

In a 2024 talk [3], it is shown Flex Attention achieves 90% of FlashAttention3 with some good tuning on H100.
