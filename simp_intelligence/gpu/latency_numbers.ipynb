In the GPU mode tvm-ffi talk by Tianqi Chen, some important latency numbers are listed (assume in a hot loop):
* native or compiled code access (~ 2-10 ns)
* Python to native code invocation (~ 0.1 us)
* Host driver overhead to launch a CUDA kernel (~ 1-2 us)
* Blackwell device overhead of a RMSNorm kernel (~ 2-3 us)
* Torch.add two numbers on CUDA (~ 2-3 us)
