## Division and Product
The (logical) division is defined as $A \oslash B := A \circ (B, \tilde B)$,
and the (logical) product, $A \otimes B := (A, \tilde A \circ B)$.

In the pseudo code, the division:
```python
compl = other.complement(self.size())
cat = Layout.from_concate(other, compl)
return self.composite(cat)
```
and for product:
```python
size = self.size() * other.cosize()
compl = self.complement(size)
compo = compl.composite(other)
return Layout.from_concate(self, compo)
```

````python
A = Layout.from_string('(4,2,3):(2,1,8)').visualize('A')
print('A.size() =', A.size())
A.logical_divide(Layout.from_string('4:2').visualize('B'), visualize_steps=True)
````

However, $A \oslash B \otimes B \not= A$:

````python
Layout.from_string('((2, 2), (2, 3)):((4, 1), (2, 8))').logical_product(Layout.from_string('4:2'))
````
