## Example
Given an example and its computation graph:

![](compute-graph-example.png)

We will have two ways to compute diffferentiation:

## Forward Auto-Differentiation (Forward AD)
Forward-AD first assumes the target *leaf variable* w.r.t. which we are going to take derivatives (e.g., $x_1$ below), and then build the gradients to the root in a bottom-up way:

![](forward-AD.png)

## Reverse Auto-Differentiation (Reverse AD)
On the other hand, the reverse-AD starts from the root node and going top-down where each node it tries to take derivatives w.r.t. only that node of *non-leaf variable* (i.e., $\nu_i$):

![](reverse-AD.png)
